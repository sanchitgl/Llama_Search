```python
dataset_path = "../datasets/msmarco_tiny/"
corpus_file = "collection.tsv"
qrels_test_file = "qrels.dl20-passage.txt"
```

```python
import collections
import pytrec_eval
import json

def load_corpus_tsv(path):
    """Returns a dictionary whose keys are passage ids and values are passage texts."""
    corpus = {}
    with open(path) as f:
        for line in f:
            passage_id, passage_text = line.strip().split('\t')
            corpus[passage_id] = {'text': passage_text}
    return corpus


def load_qrels(path):
    with open(path, 'r') as f_qrel:
        qrels = pytrec_eval.parse_qrel(f_qrel)

    return qrels

qrels = load_qrels(f"{dataset_path}{qrels_test_file}")
```

```python
shorter_lst = set()
for k,v in qrels.items():
    if type(v) == dict:
        shorter_lst.update(list(v.keys()))
    else:
        print(k)
```

```python
corpus = load_corpus_tsv(f"{dataset_path}{corpus_file}")

key_lst = list(corpus.keys())
len(key_lst)
```

```python
tiny_size = 500_000

import random

extra_docs = random.sample(key_lst, tiny_size)
shorter_lst.update(extra_docs)
# Now shorter_lst has all docs in qrels and 500k extra docs

corpus_tiny = {}
for i in shorter_lst:
    corpus_tiny[i] = corpus[i]

import json

with open(f"{dataset_path}tiny_collection.json", 'w') as f:
    json.dump(corpus_tiny, f)
```
